```python
import pandas as pd

train = pd.read_csv('kaggle9/train.csv')
test = pd.read_csv('kaggle9/test.csv')
```

```python
train.describe()
```

```python
from sklearn.model_selection import train_test_split

#x_train, x_test, y_train, y_test = train_test_split(train.drop('id','claim'))
```

```python
x_train = train.drop(['id','claim'], axis = 1)
y_train = train['claim']
x_test = test.drop('id', axis = 1)
```

```python
train['f1'] = train['f1'].fillna(train['f1'].mean())
features = x_train.columns.values.tolist()
features
```

```python
for feature in features:
    x_train[feature] = x_train[feature].fillna(x_train[feature].mean())
    x_test[feature] = x_test[feature].fillna(x_train[feature].mean())
```

```python
from sklearn.tree import DecisionTreeClassifier
import numpy as np
dt_clf = DecisionTreeClassifier()
dt_clf.fit(x_train, y_train)
pred = dt_clf.predict(x_test)
pred
```

```python
len(pred)
```

```python
test = pd.read_csv('kaggle9/test.csv')
```

```python
test['claim'] = pred
final = test[['id', 'claim']]
```
