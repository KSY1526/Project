# 범주형 변수 전처리

```python
codes = ['신고세관부호', '수입거래구분코드', '수입종류코드', '징수형태코드', '운송수단유형코드', '관세율구분코드', '원산지표시유무코드']

def outputs(names):
    print(names)
    cnt = 0 
    for i in list(train[names].value_counts().index):
        print(i, np.round(train['우범여부'][train[names] == i].mean(),3), len(train[train[names] == i]))
        cnt += 1
    print('총 개수는', cnt)
    print()

for i in codes:
    outputs(i)
```

```python
# 이름, 한계값, 대체 값
def replaces(names, value, replace):
    tem = train[names].value_counts()
    tem = tem[tem >= value].index
    train[names] = train.apply(lambda x : x[names] if (x[names] in tem) else replace, axis = 1)
    public[names] = public.apply(lambda x : x[names] if (x[names] in tem) else replace, axis = 1)
    private[names] = private.apply(lambda x : x[names] if (x[names] in tem) else replace, axis = 1)


    return

# 이름, 변환 전 값, 대체 값
def replaces2(names, value, replace):
    train[names][train[names] == value] = replace
    public[names][public[names] == value] = replace
    private[names][private[names] == value] = replace
    return

# 20개 이하의 부호를 전부 99로 대체
replaces('신고세관부호', 35, 99)
replaces('수입거래구분코드', 30, 99)
replaces('수입종류코드', 30, 99)
replaces('징수형태코드', 30, 99)
replaces('관세율구분코드', 19, 'Z')

replaces2('운송수단유형코드', 20, 30) # 철도와 차량이 유사하다고 판단.
replaces2('원산지표시유무코드', 'N', 'G') # 동일한 원산지 표시 대상

for i in codes:
    outputs(i)
```

```python
train.fillna('NULL', inplace = True)
public.fillna('NULL', inplace = True)
private.fillna('NULL', inplace = True)

replaces('신고인부호', 100, 'ABCDEF')
replaces('수입자', 50, 'ABCDEF')
replaces('해외거래처부호', 50, 'ABCDEF')
replaces('특송업체부호', 50, 'ABCDEF')
```

```python
peoples = ['신고인부호', '수입자', '해외거래처부호', '특송업체부호']

for i in peoples:
    outputs(i)
```

```python
nations = ['적출국가코드', '원산지국가코드']

for i in nations:
    outputs(i)
```

```python
RnationGood = ['KH','ZZ','CZ', 'AR','EE','GT','UA','BG','MT','SI', 'CL']
RnationBad = ['AZ', 'RO', 'SK', 'QA', 'FI']

def Rnation(x):
    if (x['적출국가코드'] in RnationGood):
        return 'Good'

    elif (x['적출국가코드'] in RnationBad):
        return 'Bad'
    
    else:
        return x['적출국가코드']

train['적출국가코드'] = train.apply(lambda x : Rnation(x), axis = 1)
public['적출국가코드'] = public.apply(lambda x : Rnation(x), axis = 1)
private['적출국가코드'] = private.apply(lambda x : Rnation(x), axis = 1)


replaces('적출국가코드', 50, 'ZZZ')
outputs('적출국가코드')
```

```python
OnationGood = ['UA', 'NZ', 'FI', 'CL', 'LT', 'AE', 'AR', 'EE', 
               'CR', 'GT', 'ZZ', 'MD', 'PR', 'DO', 'UZ', 'SA',' EC']

def Onation(x):
    if (x['원산지국가코드'] in OnationGood):
        return 'Good'

    else:
        return x['원산지국가코드']

train['원산지국가코드'] = train.apply(lambda x : Onation(x), axis = 1)
public['원산지국가코드'] = public.apply(lambda x : Onation(x), axis = 1)
private['원산지국가코드'] = private.apply(lambda x : Onation(x), axis = 1)


replaces('원산지국가코드', 50, 'ZZZ')
outputs('원산지국가코드')
```

```python
train['HS10단위부호'] = train['HS10단위부호'].astype('str')
public['HS10단위부호'] = public['HS10단위부호'].astype('str')
private['HS10단위부호'] = private['HS10단위부호'].astype('str')

train['HS10단위부호'] = train.apply(lambda x : x['HS10단위부호'][:2], axis = 1)
public['HS10단위부호'] = public.apply(lambda x : x['HS10단위부호'][:2], axis = 1)
private['HS10단위부호'] = private.apply(lambda x : x['HS10단위부호'][:2], axis = 1)


outputs('HS10단위부호')
```

```python
HGood = ['37', '51', '80', '26', '75', '50', '93', '17', '16']

def Hcode(x):
    if (x['HS10단위부호'] in HGood):
        return '100'

    else:
        return x['HS10단위부호']

train['HS10단위부호'] = train.apply(lambda x : Hcode(x), axis = 1)
public['HS10단위부호'] = public.apply(lambda x : Hcode(x), axis = 1)
private['HS10단위부호'] = private.apply(lambda x : Hcode(x), axis = 1)


replaces('HS10단위부호', 50, '00')
outputs('HS10단위부호')
```

```python
import datetime

train['신고일자'] = pd.to_datetime(train['신고일자'])
public['신고일자'] = pd.to_datetime(public['신고일자'])
private['신고일자'] = pd.to_datetime(private['신고일자'])

train['월'] = train['신고일자'].dt.month
public['월'] = public['신고일자'].dt.month 
private['월'] = private['신고일자'].dt.month 


outputs('월')
```

```python
train['요일'] = train['신고일자'].dt.weekday
public['요일'] = public['신고일자'].dt.weekday
private['요일'] = private['신고일자'].dt.weekday


outputs('요일')
```

```python
import holidays

kr_holidays = holidays.KR()

train['holiday'] = train['신고일자'].apply(lambda x: True if x in kr_holidays else False)
public['holiday'] = public['신고일자'].apply(lambda x: True if x in kr_holidays else False)
private['holiday'] = private['신고일자'].apply(lambda x: True if x in kr_holidays else False)


outputs('holiday')
```

```python
def weekdays(x):
    if (x['holiday'] == True):
        return 2

    elif (x['요일'] < 5):
        return 0
    else:
        return 1


train['공휴일'] = train.apply(lambda x : weekdays(x), axis = 1)
public['공휴일'] = public.apply(lambda x : weekdays(x), axis = 1)
private['공휴일'] = private.apply(lambda x : weekdays(x), axis = 1)


outputs('공휴일')
```
